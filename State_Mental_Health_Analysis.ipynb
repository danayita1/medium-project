```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python

data = pd.read_csv('mental_health_state_rank.csv')

data.head()

top10 = data.sort_values('Overall_Rank').head(10)
print("Top 10 States by Mental Health Ranking:")
print(top10)
```

```python
data['Overall_Rank'].describe()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Overall_Rank', y='State', data=top10, palette='viridis', dodge=False)
plt.title('Top 10 States by Mental Health Ranking')
plt.xlabel('Overall Rank (Lower is Better)')
plt.ylabel('State')
plt.show()
```

```python
bottom10 = data.sort_values('Overall_Rank', ascending=False).head(10)
print("\nBottom 10 States by Mental Health Ranking:")
print(bottom10)
```
